```python
large_airports_df = airport_data.loc[airport_data['type'] == 'large_airport']

world_airports = folium.Map(location=[0.00,0.00],zoom_start=2)

latitude = list(large_airports_df['latitude_deg'])
longitude = list(large_airports_df['longitude_deg'])
airports_name = list(large_airports_df['name'])

for lat, long, name in zip(latitude, longitude, airports_name):
    world_airports.add_child(folium.Marker(location=[lat,long], popup=name, icon=folium.Icon(color='green')))

world_airports
```

```python
world_airports2 = folium.Map(location=[0.00,0.00],zoom_start=2)

for lat, long, name in zip(latitude, longitude, airports_name):
    world_airports2.add_child(folium.CircleMarker(location=[lat,long], popup=name, radius=5, color='gray',fill_color='orange', fill_opacity=0.7))

world_airports2
```

```python
elevation = list(large_airports_df['elevation_ft'])

def elevation_function(elevation_value):
    if elevation_value < 1000:
        return 'green'
    elif 1000 <= elevation_value < 6000:
        return 'orange'
    else:
        return 'blue'
    
world_airports3 = folium.Map(location=[0.00,0.00],zoom_start=2)

for lat, long, name, elev in zip(latitude, longitude, airports_name, elevation):
    world_airports3.add_child(folium.CircleMarker(location=[lat,long], popup=name, radius=5, color=elevation_function(elev),fill_color=elevation_function(elev), fill_opacity=0.7))

world_airports3
```
